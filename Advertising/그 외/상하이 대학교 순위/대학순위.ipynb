```python
#이러한 순위에 레이블 인코딩에 유용한 범주를 할당

#Assigning a category to these rank that would be useful for label encoding.

#such as

#top 50 : S class.
#between 50-100 : A1 class.
#between 100-150 : A2 class.
#between 150-200 : A class.
#between 200-300 : B1 class.
#between 300-400 : B2 class.
#between 400-500 : B class.
#between 500-600 : C1 class.
#between 600-700 : C2 class.
#between 700-800 : C class.
#between 800-900 : D class.
#between 900-1000 : E class.
```

```python
for index, value in enumerate(df["Rank"]):
    if value == "901-1000":
        df.at[index, "Class"] = 'E'
    elif value == "801-900" :
        df.at[index, "Class"] = 'D'
    elif value == "701-800" :
        df.at[index, "Class"] = 'C'
    elif value == "601-700" :
        df.at[index, "Class"] = 'C2'
    elif value == "501-600" :
        df.at[index, "Class"] = 'C1'
    elif value == "401-500" :
        df.at[index, "Class"] = 'B'
    elif value == "301-400" :
        df.at[index, "Class"] = 'B2'
    elif value == "201-300" :
        df.at[index, "Class"] = 'B1'
    elif value == "151-200" :
        df.at[index, "Class"] = 'A'
    elif value == "101-150" :
        df.at[index, "Class"] = 'A2'
    elif int(value) > 50 and int(value) <= 100:
        df.at[index, "Class"] = 'A1'
    else:
        df.at[index, "Class"] = 'S'

df["Class"].unique()
        
```
